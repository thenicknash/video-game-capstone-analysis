<h1>Sales and Ratings Data Analysis</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

games_df = pd.read_csv('../data/normalized_games_data.csv')
```

<h2>1. Is there a correlation between the sales and ratings of video games in general?</h2>

<h4>First, we will look at the Metacritic scores of the games and see if there is a correlation between the scores and the sales.</h4>

```python
sales_metacritic_rating_df = games_df[['total_shipped', 'metacritic_score']]

# Drop rows with NaN values
sales_metacritic_rating_df = sales_metacritic_rating_df.dropna()

game_sales = np.array(sales_metacritic_rating_df['total_shipped'].astype(float))
game_metacritic_ratings = np.array(sales_metacritic_rating_df['metacritic_score'].astype(float))

print(f'The number of games we are analyzing for this question: {len(sales_metacritic_rating_df)}')

plt.scatter(game_sales, game_metacritic_ratings)
plt.xlabel('Game Sales (in millions)')
plt.ylabel('Metacritic Score')
plt.show()
```

```python
correlation_matrix = np.corrcoef(game_sales, game_metacritic_ratings)
correlation_coefficient = correlation_matrix[0,1]

print(f'The correlation between video game sales and Metacritic scores: {correlation_coefficient}')
```

<h4>As we can see, the correlation between volume of sales and Metacritic ratings is extremely low. This means that there is essentially no correlation between the two data points, and that the volume of sales of a video game does not depend on the Metacritic rating.</h4>

<h4>Next, we will look at the correlation between the volume of sales and the user scores.</h4>

```python
sales_metacritic_user_rating_df = games_df[['total_shipped', 'metacritic_user_score']]

# Drop rows with NaN values
sales_metacritic_user_rating_df = sales_metacritic_user_rating_df.dropna()

game_sales = np.array(sales_metacritic_user_rating_df['total_shipped'].astype(float))
game_metacritic_user_ratings = np.array(sales_metacritic_user_rating_df['metacritic_user_score'].astype(float))

print(f'The number of games we are analyzing for this question: {len(sales_metacritic_user_rating_df)}')

plt.scatter(game_sales, game_metacritic_user_ratings)
plt.xlabel('Game Sales (in millions)')
plt.ylabel('Metacritic Users Score')
plt.show()
```

```python
correlation_matrix = np.corrcoef(game_sales, game_metacritic_user_ratings)
correlation_coefficient = correlation_matrix[0,1]

print(f'The correlation between video game sales and Metacritic user scores: {correlation_coefficient}')
```

<h4>Metacritic user scores have an even lower correlation with the volume of sales than the Metacritic critic scores to the point where the correlation is negligible. This means that the volume of sales of a video game does not depend on the Metacritic user score.</h4>

<h4>Let's do one final check against the VGChartz scores.</h4>

```python
sales_vgchartz_df = games_df[['total_shipped', 'vgchartz_score']]

# Drop rows with NaN values in the vgchartz_score column
sales_vgchartz_df = sales_vgchartz_df.dropna()

game_sales = np.array(sales_vgchartz_df['total_shipped'].astype(float))
game_metacritic_user_ratings = np.array(sales_vgchartz_df['vgchartz_score'].astype(float))

print(f'The number of games we are analyzing for this question: {len(sales_vgchartz_df)}')

plt.scatter(game_sales, game_metacritic_user_ratings)
plt.xlabel('Game Sales (in millions)')
plt.ylabel('VGChartz Score')
plt.show()
```

```python
correlation_matrix = np.corrcoef(game_sales, game_metacritic_user_ratings)
correlation_coefficient = correlation_matrix[0,1]

print(f'The correlation between video game sales and VGChartz scores: {correlation_coefficient}')
```

<h4>The correlation between the volume of sales and the VGChartz scores is the highest in the ratings category, but it is still abysmal at < 0.2. The increase in the correlation rate could be attributed to the much smaller sample size at around 1,300 v. around 13,000.</h4>

<h3>CONCLUSION: There is no correlation between the volume of sales and the ratings of video games from Metacritic, Metacritic users, or VGChartz.</h3>
